```python
def upper(nums,x):
    n = len(nums)
    low = 0
    high = n-1
    lb = n
    while low<=high:
        mid=(low+high)//2
        if nums[mid]>=x:
            high = mid-1
            lb = mid
        else:
            low = mid+1
    return lb
def lower(nums,x):
    n = len(nums)
    low = 0
    high = n-1
    ub = n
    while low<=high:
        mid=(low+high)//2
        if nums[mid]>x:
            high = mid-1
        else:
            ub = mid
            low = mid+1
    return ub
nums = [1,2,3,3,3,3,3,5,6,8,9,9,10]
ub = upper(nums,3)
lb = lower(nums,3)
print("[",ub,lb,"]")

tub = upper(nums,9)
tlb = lower(nums,9)
print("[",tub,tlb,"]")
```
