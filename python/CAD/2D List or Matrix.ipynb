```python
nums=[[5,20,3],[7,10,0],[15,8,7]]
rows = len(nums)
cols = len(nums[0])
for i in range(rows):
    for j in range(cols):
        print(nums[i][j], end=' ')
    print()  
```

```python
for i in range(rows):
    for j in range(cols):
        if j>=i:
            print(nums[i][j], end=' ')
        else:
            print("*",end=' ')
    print()  
```

```python
for i in range(rows):
    for j in range(cols):
        if j<=i:
            print(nums[i][j], end=' ')
        else:
            print('*',end=' ')
    print()  
```

```python
for i in range(rows):
    for j in range(cols):
        if j==i:
            print(nums[i][j], end=' ')
        else:
            print('*',end=' ')
    print()  
```

Transpose of matrix

```python
#making 3X2 matrix
nums2 = [[5,9,1],[2,3,7]]
rows = len(nums2)
cols = len(nums2[1])
result = [[0]*2 for _ in range(3)]
for i in range(0,rows):
    for j in range(0,cols):
        result[j][i]=nums2[i][j]
print(result)
```
