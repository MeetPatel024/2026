```python
'''
[[7,9,2,3]
[20,8,0,10]
[29,0,-10,5]
[4,14,6,7]]
'''
```

```python
nums=[[7,9,2,3],
      [20,8,0,10],
      [29,0,-10,5],
      [4,14,6,7]]
print(nums)
rows=len(nums)
cols=len(nums[0])
zlist_row=[]
zlist_col=[]
for i in range(rows):
    for j in range(cols):
        if nums[i][j]==0:
            zlist_row.append(i)
            zlist_col.append(j)
print(zlist_col,zlist_row)
for i in range(rows):
    for j in range(cols):
        if i in zlist_row or j in zlist_col:
            nums[i][j]=0
    
print(nums)
```

Optimal

```python
nums=[[7,9,2,3],
      [20,8,0,10],
      [29,0,-10,5],
      [4,14,6,7]]
r = len(nums)
c = len(nums[0])
rowtrack=[0 for _ in range(r)]
coltrack=[0 for _ in range(c)]
for i in range(r):
    for j in range(c):
        if nums[i][j]==0:
            rowtrack[i]=-1
            coltrack[j]=-1
        
for i in range(r):
    for j in range(c):
        if rowtrack[i]==-1 or coltrack[j]==-1:
            nums[i][j]=0

print(nums)
```
