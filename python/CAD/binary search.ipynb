```python
def binary_search(nums,target):
    n=len(nums) 
    low = 0
    high = n-1
    while high>=low:
        mid= (low+high)//2
        if target==nums[mid]:
            return mid
        elif target<nums[mid]:
            high = mid-1
        else:
            low = mid+1
    return "not found"

nums = [2,4,6,7,9,11,18,19]
target = 19
print(binary_search(nums,target))
```

```python
def binary_search_recursion(nums,low,high,t):
    if low>high:
        return "not found"
    mid=(low+high)//2
    if t==nums[mid]:
        return mid
    elif t<nums[mid]:
        return binary_search_recursion(nums,low,mid-1,t)
    else:
        return binary_search_recursion(nums,mid+1,high,t)

nums = [2,4,6,7,9,11,18,19]
target = 4

print(binary_search_recursion(nums,0,len(nums)-1,target))
```
