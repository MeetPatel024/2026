```python
def fc(nums,target):
    n = len(nums)
    lb = n
    ub = n
    low = 0
    high = n-1
    while low<=high:
        mid = (low+high)//2
        if nums[mid]==target:
            return ([nums[mid],nums[mid]])
        if nums[mid]>=target:
            ub = mid
            lb = mid
            high = mid-1
        else:
            low = mid+1
    lis = [nums[lb-1],nums[ub]]
    return lis
nums = [3,4,4,4,8,9,9,10,12,12,14,15]

print(fc(nums,5))
```
